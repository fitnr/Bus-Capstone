## Matthew's comments
Let's first calculate the difference between the scheduled headway and actual headway.

```python
# this uses the groupby to ensure it only does the diff() between sequential members of the same group
sample['scheduled_headway'] = sample.sort_values(['stop_id','arrival_scheduled']).groupby('stop_id')['arrival_scheduled'].diff()
# note that when I do actual headway, I am NOT dealing with the missing values.  This assumes that the bus never came for those trips.
sample['actual_headway'] = sample.sort_values(['stop_id','arrival_actual_estimated']).groupby('stop_id')['arrival_actual_estimated'].diff()
```

Then calculate the difference between the two for each bus arrival.  (ewt means Excess Wait Time).

Finally, you can summary the ewt values.

```python
ewt = sample['actual_headway']-sample['scheduled_headway']
```

```python
# percentage of actual headways that are more than 5 minutes longer than scheduled headway
float(sum(ewt > timedelta(minutes=5)))/len(ewt)
```
